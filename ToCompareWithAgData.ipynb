```python
# OurVr = ourtemp.loc[:,['Variabel','Ours']]
# OurVr.head()
# len(OurVr)

OurVr = ourtemp.copy()
print(len(OurVr))
OurVr.head()
```

```python
AgFile = "AgData_Variabel.xlsx"
AgData = pd.ExcelFile(AgFile)
AgData.sheet_names
```

```python
AgTable = pd.read_excel(AgFile, sheet_name=AgData.sheet_names[2])
print(len(AgTable))
AgTable.head()
```

```python
AgVr = pd.DataFrame(columns = ['Variabel','Theirs'])
AgVr.head()
```

```python
for s in range(len(AgData.sheet_names)-1):
#     tabnm = "Our"+"_"+AgData.sheet_names[s]
    AgTable = pd.read_excel(AgFile, sheet_name=AgData.sheet_names[s])
    AgTable.loc[:,['Theirs']] = 1
    tabtemp =  AgTable.loc[:,['Variabel','Theirs']]
    AgVr = AgVr.append(tabtemp)
print(len(AgVr))
AgVr.head()
```

```python
# NetVr = pd.merge(OurVr,AgVr,on="Variabel",how="left")
# NetVr = OurVr.reset_index().merge(AgVr, how="left",on="Variabel").set_index('index')
NetVr = OurVr.merge(AgVr, how="left",on="Variabel")
print(len(NetVr))
NetVr.head()
```

```python
NetVrCopy = NetVr.copy()
NetVrCopy.drop(NetVrCopy[pd.isna(NetVrCopy.Theirs)==False].index, inplace=True)
print(len(NetVrCopy))
# NetVrCopy.head(50)
```

```python
pd.DataFrame.to_excel(NetVrCopy,"NewVarToAdd_CompareWAgData.xlsx")
```
